```python
import os
print(os.listdir(os.path.join(os.getcwd(), 'data', 's1')))
print(len(os.listdir(os.path.join(os.getcwd(), 'data', 's1'))))
print(os.listdir(os.path.join(os.getcwd(), 'data', 'alignments', 's1')))
print(len(os.listdir(os.path.join(os.getcwd(), 'data', 'alignments', 's1'))))
```

```python
with open('dataset.txt', 'w') as file:
    path = os.path.join(os.getcwd(), 'data')
    listn = os.listdir(os.path.join(path, 'alignments', 's1'))
    listn = [l.split('.')[0] for l in listn]
    delimeter = '/'
    for l in listn:
        file.write(delimeter.join(os.path.join(path, 'alignments', 's1', l + '.align\n').split('\\')))
```

```python
with open('alignments.txt', 'w') as file:
    path = os.path.join(os.getcwd(), 'data')
    listn = os.listdir(os.path.join(path, 'alignments', 's1'))
    listn = [l.split('.')[0] for l in listn]
    delimeter = '/'
    for l in listn:
        file.write(delimeter.join(os.path.join(path, 's1', l + '.mpg\n').split('\\')))
```
